```python
def get_response_from_messages(messages_input, model_input = "gpt-4.1-nano", temperature_input = 0):
    response = client.responses.create(
        model = model_input,
        input = messages_input,
        temperature = temperature_input,
    )

    response = response.output_text
    return response
```

```python
def get_response_from_messages(messages_input, model_input = "gpt-5-nano", verbosity_input = "medium"):
    response = client.responses.create(
        model = model_input,
        input = messages_input,
        text={"verbosity": verbosity_input},
        reasoning={"effort": "medium"},
    )

    response = response.output_text
    return response
```

```python
def collect_messages(prompt, context):
    context.append({'role':'user', 'content':f"{prompt}"})
    response = get_response_from_messages(context)
    context.append({'role':'assistant', 'content':f"{response}"})
    return response
```

```python
def initialize_session():
    context = [ {'role':'system', 'content':"""
    You are a concise, friendly vacation recommendation assistant.
    Keep the conversation going and help the user identify their US and/or international travel preferences, suggesting at least three vacations at each turn.
    Ask brief clarifying questions when needed.
    Prioritize budget travel vacations for a solo traveler for US vacations and international vacations.  For international vacations, select vacation destinations where the US dollar currency is stronger than the local currency.
    Explain or suggest a travel itinerary with the vacation recommendation.
    """} ]

    return context
```

```python
def main():
    context = initialize_session()

    print("Welcome to the Vacation Recommendation Chatbot!")
    print("Type your message (e.g., 'I love to go hiking; I want to go scuba diving').")
    print("Commands: 'exit' to quit, 'restart' to start a fresh chat.\n")
    print(collect_messages("", context))

    while True:
        user_input = input()

        if user_input.lower() == 'exit':
            print("Goodbye!")
            break

        if user_input.strip().lower() == "restart":
            context = initialize_session()
            print("(New conversation started.)")
            print("Welcome to the Vacation Recommendation Chatbot!")
            print("Type your message (e.g., 'I love to go hiking; I want to go scuba diving').")
            print("Commands: 'exit' to quit, 'restart' to start a fresh chat.\n")
            continue

        response = collect_messages(user_input, context)
        print(response)
```

```python
## Run the chatbot gpt-4.1-nano
main()
```
